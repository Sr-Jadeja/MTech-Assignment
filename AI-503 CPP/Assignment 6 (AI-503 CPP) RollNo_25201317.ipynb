```python
# 1) Write a Python program to create and display all combinations of letters,
# selecting each letter from a different key in a dictionary.
# Sample data: {'1':['a','b'], '2':['c','d']}
# Expected Output:
# ac
# ad
# bc
# bd

data = {'1': ['a', 'b'], '2': ['c', 'd']}

keys = list(data.keys())
print(keys)
for first in data[keys[0]]:
    for second in data[keys[1]]:
        print(first + second)
```

```python
# 2) Write a Python program to get the top three items in a shop.
# Sample data: {'item1': 45.50, 'item2':35, 'item3': 41.30, 'item4':55, 'item5':24}
# Expected Output:
# item4 55
# item1 45.5
# item3 41.3

shop = {'item1': 45.50, 'item2':35, 'item3': 41.30, 'item4':55, 'item5':24}

# Convert dictionary into a list of (key, value) pairs
items = list(shop.items())

# Manual sorting (descending) without using sorted()
for i in range(len(items)):
    for j in range(i + 1, len(items)):
        if items[j][1] > items[i][1]:
            items[i], items[j] = items[j], items[i]

# Display top 3
for i in range(3):
    print(items[i][0], items[i][1])
```

```python
# 3) Write a Python program to find the specified number of maximum values
# in a given dictionary.

data = {'a':5, 'b':14, 'c':32, 'd':35, 'e':24, 'f':100, 'g':57, 'h':8, 'i':100}

# Convert to list of (key, value)
pairs = list(data.items())

# Manual sort (descending by value)
for i in range(len(pairs)):
    for j in range(i + 1, len(pairs)):
        if pairs[j][1] > pairs[i][1]:
            pairs[i], pairs[j] = pairs[j], pairs[i]

# Function to get top n keys
def top_n(n):
    result = []
    for i in range(n):
        result.append(pairs[i][0])
    return result

print("1 maximum value(s):", top_n(1))
print("2 maximum value(s):", top_n(2))
print("5 maximum value(s):", top_n(5))
```

```python
# 4) Write a Python program to combine two lists into a dictionary.
# The elements of the first list serve as keys and the elements of the second as values.

keys = ['a', 'b', 'c', 'd', 'e', 'f']
values = [1, 2, 3, 4, 5]

# Manual combination without zip()
combined = {}
for i in range(len(values)):
    if i < len(keys):
        combined[keys[i]] = values[i]

print(combined)
```

```python
# 5) Write a Python function that evaluates mathematical expressions
# represented as dictionaries (support +, -, *, /, variables, nested expressions).

def evaluate(expr, vars={}):
    # If expr is not a dictionary, it can be a variable or a number
    if type(expr) != dict:
        if expr in vars:
            return vars[expr]
        else:
            return float(expr)

    # Otherwise it's a dictionary: {'op': '+', 'args': [x, y]}
    op = expr['op']
    a = expr['args'][0]
    b = expr['args'][1]

    # Evaluate both sides (they can be nested)
    val1 = evaluate(a, vars)
    val2 = evaluate(b, vars)

    # Perform operation
    if op == '+':
        return val1 + val2
    elif op == '-':
        return val1 - val2
    elif op == '*':
        return val1 * val2
    elif op == '/':
        return val1 / val2


# ---------------- Example ----------------
expr1 = {'op': '+', 'args': [5, 3]}                          # 5 + 3 = 8
expr2 = {'op': '*', 'args': [2, {'op': '+', 'args': [3, 4]}]} # 2 * (3 + 4) = 14
expr3 = {'op': '-', 'args': ['x', 2]}                         # x - 2
expr4 = {'op': '/', 'args': [{'op': '+', 'args': ['a', 10]}, 2]} # (a + 10) / 2

vars = {'x': 7, 'a': 6}

print(evaluate(expr1))        # 8
print(evaluate(expr2))        # 14
print(evaluate(expr3, vars))  # 5
print(evaluate(expr4, vars))  # 8.0
```
