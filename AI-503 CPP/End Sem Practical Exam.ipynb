### Q1. Write a program to read a floating point number is greater than 4.14, then add 10 to the integer.

```python
while True:
    try:
        num = float(input("Enter a floating point number: "))
        break
    except ValueError:
        print("Input is not valid, please enter float")

if num > 4.14:
    int_num = int(num)
    result = int_num + 10    
    print("Result after adding 10:", result)
else:
    print("Number is not greater than 4.14")
```

### Q3.Create a base class Account and derived classes SavingsAccount andCurrentAccount. Override a method to compute interest differently in each subclass.

```python
# Base class
class Account:
    def __init__(self, balance):
        self.balance = balance

    def compute_interest(self):
        print("Interest calculation not defined")

# Derived class: SavingsAccount
class SavingsAccount(Account):
    def compute_interest(self):
        interest = self.balance * 0.04
        return interest

# Derived class: CurrentAccount
class CurrentAccount(Account):
    def compute_interest(self):
        interest = self.balance * 0.01
        return interest

s = SavingsAccount(10000)
c = CurrentAccount(10000)

print("Savings Account Interest:", s.compute_interest())
print("Current Account Interest:", c.compute_interest())
```
