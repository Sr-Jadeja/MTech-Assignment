```python
import tensorflow as tf

# Create a tensor
x = tf.constant([[1.0, 2.0], [3.0, 4.0]])
print("Tensor x:\n", x)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(8, activation='relu', input_shape=(2,)),
    tf.keras.layers.Dense(1)
])

# Print model summary
model.summary()
```

```python
A = tf.constant([[1, 2],
                 [3, 4]])

B = tf.constant([[5, 6],
                 [7, 8]])

matmul_result = tf.matmul(A, B)
print("Matrix Multiplication Result:\n", matmul_result)
```

```python
t1 = tf.constant([1, 2, 3])
t2 = tf.constant([4, 5, 6])

add_result = tf.add(t1, t2)
print("Element-wise Addition:\n", add_result)
```

2) Define a simple neural n/w and print its structure.

```python
# Import libraries
import torch
import torch.nn as nn
import torch.optim as optim

# Training Data
X = torch.tensor([[1.0], [2.0], [3.0], [4.0]])
Y = torch.tensor([[2.0], [4.0], [6.0], [8.0]])

# Define 2-layer Sequential Model
model = nn.Sequential(
    nn.Linear(1, 4),   # Input layer → Hidden layer
    nn.ReLU(),
    nn.Linear(4, 1)    # Hidden layer → Output layer
)

# Print model structure
print("Model Structure:")
print(model)

# Loss Function and Optimizer
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Training for 30 epochs
epochs = 30

for epoch in range(epochs):

    # Forward pass
    outputs = model(X)

    # Calculate loss
    loss = criterion(outputs, Y)

    # Backward pass
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # Print loss
    print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")

# Prediction
test = torch.tensor([[5.0]])

predicted = model(test)

print("\nPredicted Result for X = 5:")
print(predicted.item())
```
